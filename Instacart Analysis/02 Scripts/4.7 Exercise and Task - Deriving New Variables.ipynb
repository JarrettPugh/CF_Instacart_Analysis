### Create user-defined function

```python
# Create function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

If the value within the “prices” column within the given row is less than or equal to 5, then return the string “Low-range product.”

Or else, if the value within the “prices” column within the given row is greater than 5 and less than or equal to 15, then return the string “Mid-range product.”

Or else, if the value of the “prices” column within the given row is greater than 15, then return the string “High-range product.”

Or else, return the string “Not enough data.”

Create a new column for this using a suitable method.


```python
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create empty list 'result' to place results from the loop
# Create for-loop with if statement

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Very busy")
  elif value == 3 or value == 4:
    result.append("Not busy")
  else:
    result.append("Regularly busy")
```

#### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”


```python
# Using value_counts to print the frequency of values for 'order_hour_of_day' column


df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Using pre-defined function loc()

# Create conditions
```

```python
# Labeling 'most_orders' if 'order_hour_of_day' is 10, 11, 14, 15, 13, or 12 (10-15) (10AM - 3PM)

df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'] <= 15) & (df_ords_prods_merged['order_hour_of_day'] > 9),
       'busiest_period_of_day'] = 'Most Orders'
```

```python
# Labeling 'Fewest Orders' if 'order_hour_of_day' is 3, 4, 2, 5, 1, or 0 (0-5) (12AM - 5 AM)

df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'] <= 5) & (df_ords_prods_merged['order_hour_of_day'] >= 0),
       'busiest_period_of_day'] = 'Fewest Orders'
```

```python
# Labeling 'Average Orders' if 'order_hour_of_day' is other (6-9, 16-23) (6AM - 9 AM, 4PM - 11PM)

df_ords_prods_merged.loc[((df_ords_prods_merged['order_hour_of_day'] <= 9) & (df_ords_prods_merged['order_hour_of_day'] >= 6)) | 
       ((df_ords_prods_merged['order_hour_of_day'] <= 23) & (df_ords_prods_merged['order_hour_of_day'] >= 16)), 
       'busiest_period_of_day'] = 'Average Orders'
```

#### 5. Print the frequency for this new column.


```python
# Using value_counts to print the frequency of values for 'busiest_period_of_day' column

df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

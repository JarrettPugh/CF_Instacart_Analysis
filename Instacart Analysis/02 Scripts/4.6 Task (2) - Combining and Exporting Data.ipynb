```python
# In order to add an indicator column when I merge the 2 dataframes, I need to drop
#  the indicator column from the df_orders_products_combined dataframe first

df_orders_products_combined = df_orders_products_combined.drop('_merge', axis=1)
df_orders_products_combined.head()
```

## Line Charts and Sampling Data

```python
# Create a list of random floating numbers between 0 and 1 of the length that is equivalent to the length of 
#  ords_prods_merge df

np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

So, to recap each piece of the split code, you need to:

-Set the seed for the random number generator.

-Create 32,404,859 random numbers (the length of your ords_prods_merge dataframe) ranging between 0 and 1.

-Code a logical test comparing each of those numbers to a value of 0.7 (either less than or greater than 0.7).

-Store the results of the test in a new list, dev, in the form of boolean values (True and False).


```python
# Split your dataframe into two samples
```

```python
# Store 70% of sample in the dataframe big

big = ords_prods_merge[dev]
```

```python
# Store 30% of sample in the dataframe small

small = ords_prods_merge[~dev]
```

```python
# Double check that the length of the 2 dataframes add up together
```

```python
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
# Reducing your samples to only those columns necessary for the chart, namely “prices” and “orders_day_of_week.”

df_2 = small[['orders_day_of_week','prices']]
```

```python
# Make line chart

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
#  Export line chart using figure.savefig() function

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_orders_dow_prices.png'))
```
